```python
#Custom Tool creation 
from langchain_core.tools import tool
```

```python
@tool
def multiply(a: int, b: int) -> int:
    """Multiplies two numbers together."""
    return a * b
```

```python
resu = multiply.invoke({"a": 5, "b": 10})
print(resu)
```
